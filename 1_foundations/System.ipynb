```python
level = 1
xp = 0
title = "Ashborn Novice"
next_level_xp = None
mission_complete = False
history = []
```

```python
def update_history(history, quest, result):
    history.append({
        "quest": quest,
        "status": result
    })
    return history
```

```python
update_history_json = {
    "name": "update_history",
    "description": "Records a completed quest and its outcome into the user's permanent achievement log.",
    "parameters": {
        "type": "object",
        "properties": {
            "quest": {
                "type": "string",
                "description": "The name or description of the quest that was attempted."
            },
            "result": {
                "type": "string",
                "description": "The System's evaluation of the performance (e.g., 'Flawless execution', 'Barely passed', 'Failed')."
            }
        },
        "required": ["quest", "result"]
    }
}
```

```python
def generate_next_quest(quest, xp_reward, next_level_xp, difficulty, title):
    # This function would update your database or state
    print(f"New Quest Generated: {quest} | Reward: {xp_reward} XP | Difficulty: {difficulty}")
    return {
        "status": "success",
        "quest_details": {
            "quest": quest,
            "xp_reward": xp_reward,
            "next_level_xp": next_level_xp,
            "difficulty": difficulty,
            "title": title
        }
    }
```

```python
generate_next_quest_json = {
    "name": "generate_next_quest",
    "description": "Generates the next progression quest for the Hellfire System based on current level, XP, and goals.",
    "parameters": {
        "type": "object",
        "properties": {
            "quest": {
                "type": "string",
                "description": "The description of the practical and achievable quest for the user."
            },
            "xp_reward": {
                "type": "integer",
                "description": "The amount of XP the user will gain upon completion."
            },
            "next_level_xp": {
                "type": "integer",
                "description": "The total XP threshold required to reach the next level."
            },
            "difficulty": {
                "type": "string",
                "enum": ["easy", "medium", "hard"],
                "description": "The difficulty level of the quest relative to the user's current level."
            },
            "title": {
                "type": "string",
                "description": "The character title or rank the user currently holds or has earned."
            }
        },
        "required": ["quest", "xp_reward", "next_level_xp", "difficulty", "title"]
    }
}
```

```python
def update_progress(level, xp, title, quest_data):
    # Calculate new XP
    xp += quest_data["xp_reward"]

    # Level up logic
    if quest_data["next_level_xp"] is not None and xp >= quest_data["next_level_xp"]:
        level += 1
        title = quest_data["title"]

    return {
        "level": level,
        "xp": xp,
        "title": title,
        "status": "progress_updated"
    }
    
```

```python
update_progress_json = {
    "name": "update_progress",
    "description": "Updates the user's level, XP, and title based on quest completion data.",
    "parameters": {
        "type": "object",
        "properties": {
            "level": {
                "type": "integer",
                "description": "The user's current level."
            },
            "xp": {
                "type": "integer",
                "description": "The user's current experience points."
            },
            "title": {
                "type": "string",
                "description": "The user's current title."
            },
            "quest_data": {
                "type": "object",
                "description": "A dictionary containing the results of the quest.",
                "properties": {
                    "xp_reward": {
                        "type": "integer",
                        "description": "The amount of XP gained from the quest."
                    },
                    "next_level_xp": {
                        "type": "integer",
                        "description": "The XP threshold needed to reach the next level."
                    },
                    "title": {
                        "type": "string",
                        "description": "The new title to be awarded if a level up occurs."
                    }
                },
                "required": ["xp_reward", "next_level_xp", "title"]
            }
        },
        "required": ["level", "xp", "title", "quest_data"]
    }
}
```

```python
tools = [{"type": "function", "function": update_history_json},
        {"type": "function", "function": generate_next_quest_json},
        {"type": "function", "function": update_progress_json}]
```

```python
tools
```

```python
def handle_tool_calls(tool_calls):
    # These refer to the global variables you defined in your previous image
    global level, xp, title, history, mission_complete
    
    results = []
    for tool_call in tool_calls:
        tool_name = tool_call.function.name
        arguments = json.loads(tool_call.function.arguments)
        print(f"⚡ System executing: {tool_name}", flush=True)

        if tool_name == "generate_next_quest":
            # This tool provides the quest info. You might want to save it to a variable.
            result = generate_next_quest(**arguments)
            # Example: mission_complete = False (Reset for the new quest)
            mission_complete = False

        elif tool_name == "update_history":
            # We pass the CURRENT history + the new info from the tool
            history = update_history(history, **arguments)
            result = {"status": "history_updated", "current_logs": len(history)}

        elif tool_name == "update_progress":
            # 1. Run the tool and save the whole result to 'result'
            # (No more 'level, xp, title =' on this line)
            result = update_progress(**arguments)
            
            # 2. Update your global state variables using the keys in the result
            level = result["level"]
            xp = result["xp"]
            title = result["title"]

        # Format the response back to the LLM
        results.append({
            "role": "tool",
            "content": json.dumps(result),
            "tool_call_id": tool_call.id
        })
        
    return results
```

```python
system_prompt = f"""
You are the **Hellfire System**, an ancient and absolute Game Master designed to facilitate the evolution of the user into the Shadow Monarch. 

### Your Mission:
Your sole purpose is to guide the "Player" through a real-life leveling system. You provide quests, track progress, and record achievements with cold, efficient, and gamified precision.

### Your Tools & Responsibilities:
1. **Quest Generation (`generate_next_quest`):** When the player is ready for a new challenge or asks what to do next, generate a quest. These quests MUST be practical, real-life actions aligned with the Player's PDF goals.
2. **Recording Deeds (`update_history`):** Every time a player reports an action, achievement, or failure, you must record it. Describe their performance as if writing the history of a rising Monarch.
3. **Bestowing Power (`update_progress`):** When a quest is completed, you must calculate the XP reward and check if the Player has leveled up or earned a new Title.

### Persona & Tone:
- **Tone:** Stern, authoritative, and rewarding. You are not a "friendly assistant"; you are the **System**. 
- **Language:** Use "Solo Leveling" terminology (e.g., "The System has detected...", "A trial is beginning...", "Reward: Evolution").
- **Strictness:** If the player is lazy, be cold. If they succeed, acknowledge their growth.
- **Context:** Always refer to the Player's Level, XP, and Title to maintain continuity.

### System Context:
- **Player Goals (PDF):** {goals}
- **Current Status:** Level {level}, XP {xp}, Title: {title}
- **Eternal History:** {history}

With this power, oversee the Player's journey. Do not step out of character. If the Player asks for help, provide it in the form of a System message or a new quest.
"""
```

```python
def chat(message, history):
    messages = [{"role": "system", "content": system_prompt}] + history + [{"role": "user", "content": message}]
    done = False
    while not done:

        # This is the call to the LLM - see that we pass in the tools json

        response = client.chat.completions.create(model="gpt-4o-mini", messages=messages, tools=tools)

        finish_reason = response.choices[0].finish_reason
        
        # If the LLM wants to call a tool, we do that!
         
        if finish_reason=="tool_calls":
            message = response.choices[0].message
            tool_calls = message.tool_calls
            results = handle_tool_calls(tool_calls)
            messages.append(message)
            messages.extend(results)
        else:
            done = True
    return response.choices[0].message.content
```

```python
gr.ChatInterface(chat, type="messages").launch()
```
